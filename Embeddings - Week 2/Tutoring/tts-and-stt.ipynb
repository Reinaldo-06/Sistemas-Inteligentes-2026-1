```python
from transformers import pipeline

# Create transcription pipeline
asr = pipeline("automatic-speech-recognition", model="openai/whisper-base")

audio_url = "https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/mlk.flac"

result = asr(audio_url)
print("Transcription:", result["text"])
```

```python
from TTS.api import TTS
from IPython.display import Audio

# Load TTS model
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False, gpu=False)

# Generate audio
tts.tts_to_file(text="Hello, this is a text-to-speech test running on Kaggle.", file_path="output.wav") #TTS Text to Speech

# Play audio
Audio("output.wav")
```
