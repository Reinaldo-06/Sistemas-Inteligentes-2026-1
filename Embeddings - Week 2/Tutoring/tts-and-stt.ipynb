```python
import os

# Avoid TensorFlow/Keras conflicts and force PyTorch backend.
os.environ["USE_TF"] = "0"

from transformers import pipeline

asr = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-base",
    framework="pt",
)

audio_url = "https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/mlk.flac"
result = asr(audio_url)
print("Transcription:", result["text"])
```

```python
from gtts import gTTS
from IPython.display import Audio

text = "Hello, this is a text-to-speech test running in this notebook."
tts = gTTS(text=text, lang="en")
tts.save("output.mp3")

Audio("output.mp3")
```
